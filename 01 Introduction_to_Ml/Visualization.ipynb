## Matplot and Seaborn
 ### Why Matplot and Seaborn?
both libararies are kniwn among poplar choice for visualization in python
- matplot: known for flexibility,matlpot allows detailed customixation od visual elements making it tailored plots
- Seaborn: Built on top of Matplot, seaborn simplifies visualization tasks by offering high-level functions to create asthetically pleasing statistical pleasing stastistical graphics

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
#Line Plot
x= np.linspace(0, 10, 100)
y= np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
x= np.linspace(0, 10, 100)
y= np.sin(x)
plt.plot(x, y, label="sine wave")
plt.title("Sine wave line plot")
plt.xlabel(" value of x")
plt.ylabel("sin(x) ")
plt.legend()
plt.show()
```

```python
# Bar Chart
categories = ['A', 'B', 'C']
values = [50, 100, 75]
plt.bar(categories, values, color=['red', 'green', 'blue'])
plt.title("Bar Chart example")
plt.xlabel("categories")
plt.ylabel("values")
plt.show()
```

```python
# Histogram
# name =['A', 'B', 'C' ]
# data = [1, 2, 3]
# plt.hist(data)
# plt.show()
student_scores = np.random.randint(50, 100, 100)
plt.hist(student_scores)
plt.title("Student Score")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.show()
```

```python
#histogram and KDE plot
data = np.random.randn((1000))
sns.histplot(data, kde=True, color='green' )
plt.title("Histogram with KDE")
plt.show()
```

```python
# hitmap
matrix = np.random.rand(5,5)
sns.heatmap(matrix, annot= True, cmap= "viridis")
plt.title("Hitmap Example")
plt.show()
```

```python
# Pair PLot
sns.set_style("whitegrid")
df = sns.load_dataset("iris")
sns.pairplot(df, hue="species", markers=["o", "s", "d"])
plt.show()
```
