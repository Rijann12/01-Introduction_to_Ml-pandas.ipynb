## Replacing NaN and N/a with Unknown and Nothing respectively

```python
# fillna le chai NaN values haru sab lai unknown le replace garxa
# plus hamle "N/A" lai replace garna chai .replace use garna parxa N.A strings hoo so
df = df.fillna("Unknown").replace("N/a", "Nothing")
df
```

## Chnaging format of Phone_Number

```python
# changing the phone_number format
# suru ma .astype(str) le phone_number ma vako sab lai string banako
# ani kunai pani '\', '/', lai replace garne with nothing ''
# str[-10] le chai vako value sab lai only 10 samma rakhxa if more than 10 then it only keeps
# last 10 number and if its too short it becomes null

df['Phone_Number']= df['Phone_Number'].astype(str).str.replace(r'\D', '', regex = True).str[-10:]
```

```python
df
```

## Changing in Last_Name section

```python
# hamro data chai "/White" vanne thyo last_name ma so hamle tyo agadi aako remove garna paryo
# so we put '^/' which basically only finds at beginning and replace it with nothing ''
df['Last_Name'] = df['Last_Name'].replace('^/', '', regex= True)
df
```

```python
df
```

## Formatting Address with proper format

```python
# making the address into correct format 'location,city name'
# we used .str.split(',') : str is like accessor that lets you apply strings to each item in 'address' columns
# .split(',') le chai splits address in the column into two part based on comma like ("123 Shire Lane,Shire" yo chai "123 Shire Lane Shire" yo hunca)
# ra hamro .str[0] le chai after splitting selects first part of address i.e address name  like "123 Shire Lane" matra without city name
# ra hamro .str[1].fillna(',Unknown City') le chai after splitting selects after comma ko lai city vanera ( like '123 Shire Lane ,Shire' ma comma paxi ko 'Shire' lai city vanera select garxa)
# ra if city xaina vane or comma xaina vane hamlr 'Unknown City' chai replcae garxa
df['Address'] = df['Address'].str.split(',').str[0]+df['Address'].str.split(',').str[1].fillna(',Unknown City')
df
```

## Replacing Y,N with Yes,No respectively in Paying Customer

```python
# replacing 'Y' and 'N' with Yes and No with using .replace and putting it into dictionary
df['Paying Customer']= df['Paying Customer'].replace({'Y': 'Yes', 'N': 'No'})
df
```

## Same  change as above in Do_Not_Contact	part too

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace({'Y': 'Yes', 'N':'No'})
df
```
